```python
import d3d
import matplotlib.pyplot as plt
```

```python
hyd = d3d.DelwaqHisFile(r'sc_outflows.his')
```

```python
hyd.dates
```

```python
hyd.locs
```

```python
hyd.subs
```

Here you see how to extract data as an array of [variable, location, time] 

```python
flow1 = hyd['Flow OLake-ds (m3/s)', '9600245',:]
flow2 = hyd['Flow OLake-ds (m3/s)', hyd.locs[500], :]
```

```python
plt.plot(hyd.dates, flow1)
plt.plot(hyd.dates, flow2)

plt.xlabel('date')
plt.ylabel('flow [m3/s]')
```
